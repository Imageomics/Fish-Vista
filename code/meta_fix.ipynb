```python
import pandas as pd
```

# Fish-Vista Metadata Fix (for Dataset Viewer)

Dataset: https://huggingface.co/datasets/imageomics/fish-vista

```python
classification_csvs = ["https://huggingface.co/datasets/imageomics/fish-vista/resolve/a9464827ff5c608f4079d24d2c24b939b036aa4e/classification_test.csv",
                       "https://huggingface.co/datasets/imageomics/fish-vista/resolve/a9464827ff5c608f4079d24d2c24b939b036aa4e/classification_train.csv",
                       "https://huggingface.co/datasets/imageomics/fish-vista/resolve/a9464827ff5c608f4079d24d2c24b939b036aa4e/classification_val.csv"]

identification_csvs = ["https://huggingface.co/datasets/imageomics/fish-vista/resolve/a9464827ff5c608f4079d24d2c24b939b036aa4e/identification_test_insp.csv",
                       "https://huggingface.co/datasets/imageomics/fish-vista/resolve/a9464827ff5c608f4079d24d2c24b939b036aa4e/identification_test_lvsp.csv",
                       "https://huggingface.co/datasets/imageomics/fish-vista/resolve/a9464827ff5c608f4079d24d2c24b939b036aa4e/identification_train.csv",
                       "https://huggingface.co/datasets/imageomics/fish-vista/resolve/a9464827ff5c608f4079d24d2c24b939b036aa4e/identification_val.csv"]

segmentation_csvs = ["https://huggingface.co/datasets/imageomics/fish-vista/resolve/a9464827ff5c608f4079d24d2c24b939b036aa4e/segmentation_test.csv",
                     "https://huggingface.co/datasets/imageomics/fish-vista/resolve/a9464827ff5c608f4079d24d2c24b939b036aa4e/segmentation_train.csv",
                     "https://huggingface.co/datasets/imageomics/fish-vista/resolve/a9464827ff5c608f4079d24d2c24b939b036aa4e/segmentation_val.csv"]
```

```python
classification_df = pd.read_csv(classification_csvs[0])
classification_df.head()
```

```python
classification_df["split"] = "test"
```

```python
classification_df.shape
```

```python
def add_split(df, csv, split):
    print(f"Adding split {split}")
    temp_df = pd.read_csv(csv)
    temp_df["split"] = split
    print(temp_df.shape)
    combined_df = pd.concat([df, temp_df], ignore_index = True)
    print(combined_df.shape)
    return combined_df
```

```python
class_df = add_split(classification_df, classification_csvs[1], "train")
classification_df = add_split(class_df, classification_csvs[2], "val")
```

### Save Classification Metadata File

```python
classification_df["subset"] = "species_classification"

classification_df.to_csv("meta-subsets/classification-metadata.csv", index = False)
```

## Make Trait ID Metadata File

```python
id_df = pd.read_csv(identification_csvs[0])
id_df["split"] = "test_insp"

id_df.shape
```

```python
id_temp = add_split(id_df, identification_csvs[1], "test_lvsp")
id_temp = add_split(id_temp, identification_csvs[2], "train")
id_df = add_split(id_temp, identification_csvs[3], "val")
```

```python
id_df["subset"] = "species_trait_identification"

id_df.to_csv("meta-subsets/identification-metadata.csv", index = False)
```

## Make Segmentation Metadata File

```python
seg_df = pd.read_csv(segmentation_csvs[0])
seg_df["split"] = "test"

seg_df.shape
```

```python
seg_temp = add_split(seg_df, segmentation_csvs[1], "train")
seg_df = add_split(seg_temp, segmentation_csvs[2], "val")
```

```python
seg_df["subset"] = "trait_segmentation"

seg_df.to_csv("meta-subsets/segmentation-metadata.csv", index = False)
```

## Combine Subsets to Single Metadata File

```python
temp_combo = pd.concat([classification_df, id_df], ignore_index = True)
print(temp_combo.shape)

combo_df = pd.concat([temp_combo, seg_df], ignore_index = True)
combo_df.shape
```

```python
combo_df.to_csv("meta-subsets/metadata.csv", index = False)
```
